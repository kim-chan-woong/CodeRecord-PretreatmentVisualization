```python
import pandas as pd
import os
from datetime import date
# 쉼표로 구분된 csv파일
df=pd.read_csv('url\filename', encoding="CP949")

d=['월','화','수','목','금','토','일']
year=[]
month=[]
day=[]
weekday=[]
time=[]
time2=[]
location=[]
시=[]
구군=[]
단속수단=[]

# 날짜 형식에서 연도, 월, 일 int형으로 추출
# 시간 컬럼을 int형으로 바꾼 후 시간대 지정
# 원본데이터의 형식에 따라 인덱싱 
for i in range(0, len(df)):
    year.append(int(df.values[i][0].replace('-','')[0:4]))
    month.append(int(df.values[i][0].replace('-','')[4:6]))
    day.append(int(df.values[i][0].replace('-','')[6:8]))
    weekday.append(d[date(int(df.values[i][0].replace('-','')[0:4]),
                          int(df.values[i][0].replace('-','')[4:6]),
                          int(df.values[i][0].replace('-','')[6:8])).weekday()])
    time.append(df.values[i][1].replace(':',''))
    location.append(df.values[i][2])
    시.append('wants')
    구군.append('wants')
    단속수단.append(df.values[i][3])

for i in range(0,len(df)):
        if int(time[i]) in range(0,100):
            time2.append("00~01")
        elif int(time[i]) in range(100,200):
            time2.append("01~02")
        elif int(time[i]) in range(200,300):
            time2.append("02~03")
        elif int(time[i]) in range(300,400):
            time2.append("03~04")
        elif int(time[i]) in range(400,500):
            time2.append("04~05")
        elif int(time[i]) in range(500,600):
            time2.append("05~06")
        elif int(time[i]) in range(600,700):
            time2.append("06~07")           
        elif int(time[i]) in range(700,800):
            time2.append("07~08")
        elif int(time[i]) in range(800,900):
            time2.append("08~09")
        elif int(time[i]) in range(900,1000):
            time2.append("09~10")
        elif int(time[i]) in range(1000,1100):
            time2.append("10~11")
        elif int(time[i]) in range(1100,1200):
            time2.append("11~12")
        elif int(time[i]) in range(1200,1300):
            time2.append("12~13")
        elif int(time[i]) in range(1300,1400):
            time2.append("13~14")
        elif int(time[i]) in range(1400,1500):
            time2.append("14~15")
        elif int(time[i]) in range(1500,1600):
            time2.append("15~16")
        elif int(time[i]) in range(1600,1700):
            time2.append("16~17")
        elif int(time[i]) in range(1700,1800):
            time2.append("17~18")
        elif int(time[i]) in range(1800,1900):
            time2.append("18~19")
        elif int(time[i]) in range(1900,2000):
            time2.append("19~20")
        elif int(time[i]) in range(2000,2100):
            time2.append("20~21")
        elif int(time[i]) in range(2100,2200):
            time2.append("21~22")
        elif int(time[i]) in range(2200,2300):
            time2.append("22~23")
        elif int(time[i]) in range(2300,2400):
            time2.append("23~24")

result=pd.DataFrame({'시':시,
                     '구군':구군,
                     '연도':year,
                     '월':month,
                     '일':day,
                     '요일':weekday,
                     '시간대':time2,
                     '장소':location,
                     '단속수단':단속수단,
                     })
result.to_csv('url\want filename', encoding='CP949')
```
